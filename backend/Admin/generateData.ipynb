```python
BASE_IMAGE_URL = "https://example.com/products"

category_products = {
    1: ["Banana", "Apple", "Mango", "Orange", "Strawberry", "Kiwi", "Cherry", "Papaya", "Palm", "Coconut", "Guava"],
    2: ["Tomato", "Carrot", "Spinach", "Broccoli", "Corn", "Cabbage", "Cucumber", "Potato", "Pepper", "Onion", "Beat", "Corn"],
    3: ["Chicken", "Beef", "Pork", "Lamb", "Fish", "Mutton", "Turkey", "Duck"],
    4: ["Bread", "Croissant", "Bagel", "Muffin", "Donut", "Swiss roll", "Pie"],
    5: ["Milk", "Cheese", "Yogurt", "Butter", "Cream", "Milk Powder"],
    6: ["Chips", "Cookies", "Crackers", "Candy", "Popcorn", "Chocolate", "Cereal"],
    7: ["Juice", "Soda", "Water", "Tea", "Beer", "Wine", "Coffee", "Coke", "Energy Drink"]
}

def generate_products_all_categories():
    used_names = set()
    products_list = []

    for category_id, product_names in category_products.items():
        unique_product_names = list(set(product_names))  # remove duplicates
        for base_name in unique_product_names:
            name = base_name  # unique already
            used_names.add(name)

            image_name = name.lower().replace(" ", "_")
            description = f"Fresh and high-quality {base_name} sourced directly from trusted suppliers."

            product = (
                name,
                random.randint(20, 100),
                random.choice([True, False]),
                round(random.uniform(1.0, 20.0), 2),
                description,
                f"{BASE_IMAGE_URL}/{image_name}.jpg",
                category_id
            )
            products_list.append(product)

    return products_list
```
